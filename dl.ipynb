```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras import models
from tensorflow.keras.models import Sequential
from tensorflow.keras.regularizers import l2
```

```python
df  = pd.read_csv(r"C:\Users\gupta\OneDrive\Desktop\My Projects\Car accident Dataquiest\Dataquest_dataset.csv")
df
```

```python
df['Gender'] = df['Gender'].replace({
   'Male':1,
   'Female':0 
})

df['Alcohol_Consumption']  = df['Alcohol_Consumption'].replace({
  'Yes':1,
  'No':0 })

df['Seatbelt_Usage'] = df['Seatbelt_Usage'].replace({'Yes':1,
  'No':0 })

df['Vehicle_Type'] = df['Vehicle_Type'].replace({
  'Heavy Vehicle':6,
  'Car':4,
  'T.W':2  
})

df['Road_Type'] = df['Road_Type'].replace(
    {
      'Urban':1,
       'Rural':2 
    }
)
df['Crash_Type'] = df['Crash_Type'].replace(
    {
      'Rear-end':1,
       'Head-on':2 
    }
)
df['Road_Surface_Condition'] = df['Road_Surface_Condition'].replace(
    {
      'Icy':1,
       'Dry':2 ,
       'Wet':3
    }
)
df['Crash_Severity']=  df['Crash_Severity'].replace({
  'Minor injury':0,
   'Fatal crash':1,
    'Major injury':2
})
```

```python
df = df[['Crash_Severity','Vehicle_Speed','Age','Lane_Width','Road_Type','Alcohol_Consumption','Crash_Type','Seatbelt_Usage','Speed_Limit']]
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
x.shape
```

```python
model = Sequential()
model.add(Dense(256 , activation='relu' , input_shape=(8,) , kernel_regularizer=l2(0.0007)))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(256 , activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(128 , activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(128 , activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(64 , activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(64 , activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(32 , activation='relu'))
model.add(BatchNormalization())

model.add(Dense(32 , activation='relu'))
model.add(BatchNormalization())


model.add(Dense(3 ,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss = "mse" , optimizer='adam' , metrics=['accuracy'] )
```

```python
result = model.fit(x,y, epochs=60 , validation_split= 0.2 , batch_size=8 )
```

```python
model.save('accident.h5')
```

```python
# Test Function Definition
def test_function(query, model):
    gender_map = {'Male': 1, 'Female': 0}
    alcohol_map = {'Yes': 1, 'No': 0}
    seatbelt_map = {'Yes': 1, 'No': 0}
    vehicle_type_map = {'Heavy Vehicle': 6, 'Car': 4, 'T.W': 2}
    road_type_map = {'Urban': 1, 'Rural': 2}
    crash_type_map = {'Rear-end': 1, 'Head-on': 2}
    road_surface_map = {'Icy': 1, 'Dry': 2, 'Wet': 3}
    severity_map = {0: 'Minor injury', 1: 'Fatal crash', 2: 'Major injury'}
    
    preprocessed_query = {
        'Vehicle_Speed': query['Vehicle_Speed'],
        'Age': query['Age'],
        'Lane_Width': query['Lane_Width'],
        'Road_Type': road_type_map[query['Road_Type']],
        'Alcohol_Consumption': alcohol_map[query['Alcohol_Consumption']],
        'Crash_Type': crash_type_map[query['Crash_Type']],
        'Seatbelt_Usage': seatbelt_map[query['Seatbelt_Usage']],
        'Speed_Limit': query['Speed_Limit']
    }
    input_df = pd.DataFrame([preprocessed_query])
    
    prediction_probs = model.predict(input_df)
    max_index = np.argmax(prediction_probs)
    predicted_category = severity_map[max_index]
    return predicted_category
```

```python
query_input = {
    'Crash_Severity': 'Minor injury',
    'Vehicle_Speed': 45,
    'Crash_Time': 14,
    'Age': 30,
    'Gender': 'Male',
    'Vehicle_Type': 'Car',
    'Number_of_Lanes': 3,
    'Lane_Width': 3.5,
    'Road_Type': 'Urban',
    'Alcohol_Consumption': 'No',
    'Crash_Type': 'Rear-end',
    'Seatbelt_Usage': 'Yes',
    'Speed_Limit': 60,
    'Road_Surface_Condition': 'Dry'
}
```

```python
prediction_result = test_function(query_input, model)
print(f"Predicted Crash Severity: {prediction_result}")
```
